```python
PROCESSED_DIR = '/content/dataset_processed'
if os.path.exists(PROCESSED_DIR): shutil.rmtree(PROCESSED_DIR)
os.makedirs(PROCESSED_DIR, exist_ok=True)
IMG_SIZE = 224

def load_image(path):
    img = cv2.imread(str(path))
    if img is None:
        try:
            pil_img = Image.open(str(path)).convert('RGB')
            img = cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2BGR)
        except: return None
    return img

def apply_clahe_grayscale(img_bgr):
    gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    eq = clahe.apply(gray)
    return cv2.merge([eq, eq, eq])

def try_face_crop(img_bgr, margin=0.3):
    cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
    faces = cascade.detectMultiScale(gray, 1.1, 4, minSize=(30, 30))
    if len(faces) > 0:
        x, y, fw, fh = sorted(faces, key=lambda f: f[2]*f[3], reverse=True)[0]
        p_x, p_y = int(fw * margin), int(fh * margin)
        h, w = img_bgr.shape[:2]
        crop = img_bgr[max(0, y-p_y):min(h, y+fh+p_y), max(0, x-p_x):min(w, x+fw+p_x)]
        return crop, True
    return img_bgr, False

def preprocess_image(img_path, save_path):
    img = load_image(img_path)
    if img is None: return False, 'load_failed'

    # Decide cropping logic based on folder name
    if 'face' in str(img_path).lower():
        processed_img, face_found = try_face_crop(img)
        status = 'face' if face_found else 'centercrop'
    else:
        # Front/Back logic: Pad to Square to keep body proportions
        h, w = img.shape[:2]
        side = max(h, w)
        pad = np.zeros((side, side, 3), dtype=np.uint8)
        pad[(side-h)//2:(side-h)//2+h, (side-w)//2:(side-w)//2+w] = img
        processed_img = pad
        status = 'body_padded'

    final = apply_clahe_grayscale(processed_img)
    final = cv2.resize(final, (IMG_SIZE, IMG_SIZE), interpolation=cv2.INTER_LANCZOS4)
    cv2.imwrite(str(save_path), final)
    return True, status

def run_preprocessing(img_list, class_name):
    dst = os.path.join(PROCESSED_DIR, class_name)
    os.makedirs(dst, exist_ok=True)
    count = 0
    for idx, path in enumerate(img_list):
        ok, _ = preprocess_image(path, os.path.join(dst, f'{class_name}_{idx:04d}.jpg'))
        if ok: count += 1
    print(f'  [{class_name}] Processed {count}/{len(img_list)}')

print('🔄 Preprocessing Multi-View Dataset...')
run_preprocessing(mal_images, 'malnourished')
run_preprocessing(hlt_images, 'healthy')
```

```python
def show_samples(class_name, n=3):
    proc_path = Path(PROCESSED_DIR) / class_name
    imgs = sorted(list(proc_path.glob('*.jpg')))[:n]
    plt.figure(figsize=(10, 4))
    for i, img_p in enumerate(imgs):
        plt.subplot(1, n, i+1)
        plt.imshow(cv2.imread(str(img_p)))
        plt.title(f'{class_name} sample')
        plt.axis('off')
    plt.show()

show_samples('malnourished')
show_samples('healthy')
```

```python
TRAIN_DIR, VAL_DIR, TEST_DIR = '/content/split/train', '/content/split/val', '/content/split/test'
for cls in ['malnourished', 'healthy']:
    imgs = list((Path(PROCESSED_DIR)/cls).glob('*.jpg'))
    random.seed(42); random.shuffle(imgs)
    n = len(imgs)
    n_tr, n_val = int(n*0.7), int(n*0.15)
    splits = {'train': imgs[:n_tr], 'val': imgs[n_tr:n_tr+n_val], 'test': imgs[n_tr+n_val:]}
    for s_name, s_imgs in splits.items():
        dst = Path(f'/content/split/{s_name}') / cls
        dst.mkdir(parents=True, exist_ok=True)
        for img in s_imgs: shutil.copy(img, dst/img.name)
```

```python
train_gen = ImageDataGenerator(preprocessing_function=preprocess_input, rotation_range=20, horizontal_flip=True, brightness_range=[0.8, 1.2]).flow_from_directory(TRAIN_DIR, target_size=(224,224), batch_size=8, class_mode='binary')
val_gen = ImageDataGenerator(preprocessing_function=preprocess_input).flow_from_directory(VAL_DIR, target_size=(224,224), batch_size=8, class_mode='binary', shuffle=False)
test_gen = ImageDataGenerator(preprocessing_function=preprocess_input).flow_from_directory(TEST_DIR, target_size=(224,224), batch_size=1, class_mode='binary', shuffle=False)
CLASS_NAMES = list(train_gen.class_indices.keys())
```

```python
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224,224,3))
base_model.trainable = False
x = GlobalAveragePooling2D()(base_model.output)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(1, activation='sigmoid')(x)
model = Model(inputs=base_model.input, outputs=output)
```

```python
model.compile(optimizer=Adam(1e-3), loss='binary_crossentropy', metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])
history1 = model.fit(train_gen, epochs=20, validation_data=val_gen, verbose=1)
```

```python
base_model.trainable = True
for layer in base_model.layers[:-30]: layer.trainable = False
model.compile(optimizer=Adam(1e-5), loss='binary_crossentropy', metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])
history2 = model.fit(train_gen, epochs=20, validation_data=val_gen, verbose=1)
```

```python
# --- ADD THIS FIX AT THE TOP ---
# This ensures the model knows 0 = healthy and 1 = malnourished
if 'train_gen' in globals():
    idx_to_class = {v: k for k, v in train_gen.class_indices.items()}
else:
    # Manual fallback if train_gen is missing
    idx_to_class = {0: 'healthy', 1: 'malnourished'}

# --- REST OF YOUR ORIGINAL CODE ---
test_gen.reset()
y_true     = test_gen.classes
y_pred_prob = model.predict(test_gen, verbose=1).flatten()
y_pred      = (y_pred_prob >= 0.5).astype(int)

print('\n📋 CLASSIFICATION REPORT:')
# We use .get() to avoid further errors if the indices are unexpected
target_names = [idx_to_class.get(0, 'Class 0'), idx_to_class.get(1, 'Class 1')]
print(classification_report(y_true, y_pred, target_names=target_names))

cm_arr = confusion_matrix(y_true, y_pred)
tn, fp, fn, tp = cm_arr.ravel()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Confusion Matrix
sns.heatmap(cm_arr, annot=True, fmt='d', cmap='Blues',
            xticklabels=target_names,
            yticklabels=target_names, ax=axes[0], linewidths=1)
axes[0].set_title('Confusion Matrix', fontweight='bold')
axes[0].set_xlabel('Predicted'); axes[0].set_ylabel('True')
for i, row in enumerate([['TN','FP'],['FN','TP']]):
    for j, lbl in enumerate(row):
        axes[0].text(j+0.5, i+0.75, lbl, ha='center', color='red', fontweight='bold')

# ROC Curve
fpr, tpr, _ = roc_curve(y_true, y_pred_prob)
roc_auc     = auc(fpr, tpr)
axes[1].plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.3f}')
axes[1].plot([0,1],[0,1], 'navy', lw=1, linestyle='--', label='Random')
axes[1].fill_between(fpr, tpr, alpha=0.1, color='orange')
axes[1].set_xlabel('False Positive Rate'); axes[1].set_ylabel('True Positive Rate')
axes[1].set_title('ROC Curve', fontweight='bold')
axes[1].legend(loc='lower right'); axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('/content/evaluation_results.png', dpi=150, bbox_inches='tight')
plt.show()

# Summary Metrics
sensitivity = tp/(tp+fn) if (tp+fn)>0 else 0
specificity = tn/(tn+fp) if (tn+fp)>0 else 0
precision_v = tp/(tp+fp) if (tp+fp)>0 else 0
f1_score_v  = 2*precision_v*sensitivity/(precision_v+sensitivity) if (precision_v+sensitivity)>0 else 0
accuracy_v  = (tp+tn)/(tp+tn+fp+fn)

print(f'\n📊 SUMMARY METRICS:')
print(f'  Accuracy    : {accuracy_v:.4f} ({accuracy_v*100:.1f}%)')
print(f'  Sensitivity : {sensitivity:.4f}  ← Most important! (catching real cases)')
print(f'  Specificity : {specificity:.4f}')
print(f'  Precision   : {precision_v:.4f}')
print(f'  F1 Score    : {f1_score_v:.4f}')
print(f'  AUC-ROC     : {roc_auc:.4f}')
print(f'\n  TP={tp} FP={fp} FN={fn} TN={tn}')
```

```python
def get_gradcam(model, img_array, layer='conv5_block3_out'):
    grad_model = Model([model.inputs], [model.get_layer(layer).output, model.output])
    with tf.GradientTape() as tape:
        conv_out, preds = grad_model(img_array)
        loss = preds[:, 0]
    grads       = tape.gradient(loss, conv_out)
    pooled      = tf.reduce_mean(grads, axis=(0,1,2))
    heatmap     = conv_out[0] @ pooled[..., tf.newaxis]
    heatmap     = tf.squeeze(heatmap)
    heatmap     = tf.maximum(heatmap, 0)
    if tf.reduce_max(heatmap) > 0:
        heatmap = heatmap / tf.reduce_max(heatmap)
    return heatmap.numpy()

def show_gradcam_grid(image_dir, class_name, n=4):
    class_dir = Path(image_dir) / class_name
    samples   = random.sample(list(class_dir.glob('*.jpg')), min(n, len(list(class_dir.glob('*.jpg')))))

    fig, axes = plt.subplots(len(samples), 3, figsize=(12, 4*len(samples)))
    if len(samples)==1: axes = np.array([axes])
    fig.suptitle(f'Grad-CAM — {class_name.upper()}', fontsize=13, fontweight='bold')
    for ax, col in zip(axes[0], ['Preprocessed Image', 'Grad-CAM Heatmap', 'Overlay + Prediction']):
        ax.set_title(col, fontweight='bold')

    for row, img_path in enumerate(samples):
        proc = cv2.imread(str(img_path))
        proc_rgb = cv2.cvtColor(proc, cv2.COLOR_BGR2RGB)

        tensor = preprocess_input(proc_rgb.astype(np.float32))
        tensor = np.expand_dims(tensor, 0)

        pred_prob = float(model.predict(tensor, verbose=0)[0][0])
        pred_cls  = idx_to_class[int(pred_prob >= 0.5)]
        correct   = pred_cls == class_name

        heatmap     = get_gradcam(model, tensor)
        hm_resized  = cv2.resize(heatmap, (IMG_SIZE, IMG_SIZE))
        hm_colored  = (cm.jet(hm_resized)[:,:,:3] * 255).astype(np.uint8)
        overlay     = cv2.addWeighted(proc_rgb, 0.6, hm_colored, 0.4, 0)

        axes[row,0].imshow(proc_rgb); axes[row,0].axis('off')
        axes[row,1].imshow(hm_colored); axes[row,1].axis('off')
        color = 'green' if correct else 'red'
        axes[row,2].imshow(overlay)
        axes[row,2].set_title(f'{"✅" if correct else "❌"} Pred: {pred_cls} ({pred_prob:.2f})\nTrue: {class_name}',
                               fontsize=9, color=color, fontweight='bold')
        axes[row,2].axis('off')

    plt.tight_layout()
    plt.savefig(f'/content/gradcam_{class_name}.png', dpi=150, bbox_inches='tight')
    plt.show()
    print(f'📊 Saved: /content/gradcam_{class_name}.png')

show_gradcam_grid(TEST_DIR, 'malnourished', n=4)
show_gradcam_grid(TEST_DIR, 'healthy',      n=4)
```

```python
test_gen.reset()
y_all   = test_gen.classes
p_all   = model.predict(test_gen, verbose=0).flatten()

thresholds   = np.arange(0.2, 0.8, 0.05)
results_list = []

for t in thresholds:
    yp = (p_all >= t).astype(int)
    cm_t = confusion_matrix(y_all, yp)
    if cm_t.shape == (2,2):
        tn_t,fp_t,fn_t,tp_t = cm_t.ravel()
        s = tp_t/(tp_t+fn_t) if (tp_t+fn_t)>0 else 0
        sp= tn_t/(tn_t+fp_t) if (tn_t+fp_t)>0 else 0
        pr= tp_t/(tp_t+fp_t) if (tp_t+fp_t)>0 else 0
        f1= 2*pr*s/(pr+s)    if (pr+s)>0 else 0
        ac= (tp_t+tn_t)/(tp_t+tn_t+fp_t+fn_t)
        results_list.append({'t':t,'sen':s,'spe':sp,'f1':f1,'acc':ac,'tp':tp_t,'fp':fp_t,'fn':fn_t,'tn':tn_t})

fig, ax = plt.subplots(figsize=(10,5))
ts = [r['t'] for r in results_list]
ax.plot(ts, [r['sen'] for r in results_list], 'r-o', label='Sensitivity')
ax.plot(ts, [r['spe'] for r in results_list], 'b-o', label='Specificity')
ax.plot(ts, [r['f1']  for r in results_list], 'g-o', label='F1 Score')
ax.plot(ts, [r['acc'] for r in results_list], 'k-o', label='Accuracy')
ax.axvline(0.5, color='gray', linestyle='--', alpha=0.5, label='Default (0.5)')
ax.set_xlabel('Threshold'); ax.set_ylabel('Score')
ax.set_title('Metrics vs Threshold', fontweight='bold')
ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('/content/threshold_analysis.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'\n{"Threshold":>10} {"Sensitivity":>12} {"Specificity":>12} {"F1":>8} {"Acc":>8} {"TP":>5} {"FP":>5} {"FN":>5} {"TN":>5}')
print('-'*75)
for r in results_list:
    print(f'{r["t"]:>10.2f} {r["sen"]:>12.3f} {r["spe"]:>12.3f} {r["f1"]:>8.3f} {r["acc"]:>8.3f} {r["tp"]:>5} {r["fp"]:>5} {r["fn"]:>5} {r["tn"]:>5}')

best = next((r for r in results_list if r['sen'] >= 0.85), max(results_list, key=lambda r: r['sen']))
OPTIMAL_THRESHOLD = best['t']

print(f'\n💡 RECOMMENDED THRESHOLD: {OPTIMAL_THRESHOLD:.2f}')
print(f'   Sensitivity : {best["sen"]:.3f}')
print(f'   Specificity : {best["spe"]:.3f}')
print(f'   F1 Score    : {best["f1"]:.3f}')
print(f'\n   → Use CLASSIFICATION_THRESHOLD = {OPTIMAL_THRESHOLD:.2f} in your app')
```

```python
import cv2, numpy as np, tensorflow as tf, json
from tensorflow.keras.applications.resnet50 import preprocess_input
from google.colab import files
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 1. Load your trained model and metadata
MODEL_PATH = '/content/malnutrition_resnet50.h5'
METADATA_PATH = '/content/model_metadata.json'

model = tf.keras.models.load_model(MODEL_PATH)
with open(METADATA_PATH) as f:
    metadata = json.load(f)

THRESHOLD = metadata['optimal_threshold']
CLASS_NAMES = metadata['classes']
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# --- Grad-CAM Logic ---
def get_gradcam_heatmap(model, img_array, last_conv_layer_name="conv5_block3_out"):
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(last_conv_layer_name).output, model.output]
    )
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        class_channel = preds[:, 0] # Binary sigmoid output

    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / tf.reduce_max(heatmap)
    return heatmap.numpy()

def preprocess_view(img_bgr, view_type):
    img = img_bgr.copy()
    if view_type == 'face':
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.1, 4, minSize=(30,30))
        if len(faces) > 0:
            x,y,w,h = sorted(faces, key=lambda f: f[2]*f[3], reverse=True)[0]
            pad = int(max(w,h)*0.3)
            img = img[max(0,y-pad):y+h+pad, max(0,x-pad):x+w+pad]

    h, w = img.shape[:2]
    side = max(h, w)
    canvas = np.zeros((side, side, 3), dtype=np.uint8)
    canvas[(side-h)//2:(side-h)//2+h, (side-w)//2:(side-w)//2+w] = img

    gray_final = cv2.cvtColor(canvas, cv2.COLOR_BGR2GRAY)
    eq = cv2.createCLAHE(2.0,(8,8)).apply(gray_final)
    img_3ch = cv2.merge([eq,eq,eq])
    img_res = cv2.resize(img_3ch, (224, 224))
    return np.expand_dims(preprocess_input(img_res.astype(np.float32)), 0), img_3ch

# 2. Process Views
views = ['face', 'front', 'back']
probs = []
results_data = []

print("🚀 Starting Multi-View Grad-CAM Analysis...")
for view in views:
    print(f"\n📸 Upload {view.upper()} image:")
    uploaded = files.upload()
    fn = list(uploaded.keys())[0]

    raw_img = cv2.imdecode(np.frombuffer(open(fn, 'rb').read(), np.uint8), cv2.IMREAD_COLOR)
    tensor, processed_img = preprocess_view(raw_img, view)

    p = float(model.predict(tensor, verbose=0)[0][0])
    heatmap = get_gradcam_heatmap(model, tensor)

    probs.append(p)
    results_data.append({'img': processed_img, 'heatmap': heatmap, 'prob': p, 'view': view})

# 3. Final Summary
avg_prob = np.mean(probs)
final_result = CLASS_NAMES[int(avg_prob >= THRESHOLD)]

print("\n" + "="*40)
print(f"📊 FINAL SYSTEM DIAGNOSIS: {final_result.upper()}")
print(f"📈 AVERAGE PROBABILITY: {avg_prob:.4f}")
print("="*40)

# 4. Display Grad-CAM Grid
fig, axes = plt.subplots(3, 2, figsize=(10, 15))
for i, data in enumerate(results_data):
    # Original (Processed)
    axes[i, 0].imshow(cv2.cvtColor(data['img'], cv2.COLOR_BGR2RGB))
    axes[i, 0].set_title(f"View: {data['view'].capitalize()}\nProb: {data['prob']:.2f}")
    axes[i, 0].axis('off')

    # Grad-CAM Overlay
    img_rgb = cv2.cvtColor(cv2.resize(data['img'], (224, 224)), cv2.COLOR_BGR2RGB)
    heatmap = cv2.resize(data['heatmap'], (224, 224))
    heatmap = np.uint8(255 * heatmap)
    heatmap_color = cm.jet(heatmap)[:, :, :3]
    heatmap_color = (heatmap_color * 255).astype(np.uint8)

    superimposed = cv2.addWeighted(img_rgb, 0.6, heatmap_color, 0.4, 0)
    axes[i, 1].imshow(superimposed)
    axes[i, 1].set_title("Grad-CAM Focus Area")
    axes[i, 1].axis('off')

plt.tight_layout()
plt.show()
```
